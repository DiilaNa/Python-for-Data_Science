```python
df = pd.DataFrame({
    "Product": ["Laptop","Mouse","Keyboard","Monitor","HeadtSet"],
    "UnitSold": [12,50,34,20,40],
    "Price": [120000,1300,1500,25000,6000],
    "Discount":[0.10,0.05,0.00,0.12,0.07]
})
```

```python
df
```

```python
df["Final_Price"] = df['Price']  - (df['Price'] * df['Discount'])
print(df)
```

```python
df["TotalRevenue"] = df["UnitSold"] * df["Final_Price"]
print(df)
```

```python
df[df["TotalRevenue"] > 500000]
```

```python
print(df["TotalRevenue"].sum())
```
