- when adding two arrays together it adds coressponding array elements of both arrays

```python
arr_1 = np.array([4,6,7,8,9,10])
arr_2 = np.array([3,4,6,2,8,13])

arr_3 = arr_1 + arr_2
print(arr_3)
```

```python
arr_4 = arr_1 - arr_2
print(arr_4)
```

```python
arr_5 = arr_1 * arr_2
print(arr_5)
```

```python
arr_6 = arr_2 / arr_1
print(arr_6)
```

- Numeric operations with Numpy (get the sum of elements)

```python
arr_1 = np.array([4,6,7,8,9,10])
arr_2 = np.array([3,4,6,2,8,13])

sum_of_arrays = np.sum(arr_1 + arr_2)
print(sum_of_arrays)

avaerage = np.mean(arr_1 + arr_2)
print(avaerage)

print(np.min(arr_1))
print(np.max(arr_1))
print(np.sqrt(arr_2))
```

### Array Indexing

```python
arr = np.array([[1,2,4],[3,5,6]])
print(arr)
```

```python
print(arr[1][2])
```

```python
arr_3D = np.array([
    [[1,2],[5,7],[9,8]],
    [[3,4],[6,8],[10,12]],
    [[11,13],[14,16],[18,20]]
    ])
```

- Print 14 using Indexing and Negative Indexing

```python
print(arr_3D[2,1,0])
print(arr_3D[-1,-2,-2])
```

- Print 3 using Indexing and Negative Indexing

```python
print(arr_3D[1,0,0])
print(arr_3D[-2,-3,-2])
```

```python
array = np.array([10,20,15,18,14,26,30])

con = array > 15
print(con)

filtered_arr = array[con]
print(filtered_arr)
```

- after filtering gives you 1 dimension not 2

```python
arr = np.array([
    [12,5,7,20],
    [3,15,8,2],
    [10,25,6,18],
    [4,11,30,9]
])
arr_1 = arr[((arr>10)|(arr<5)) &(arr != 15)]
print(arr_1)
```
