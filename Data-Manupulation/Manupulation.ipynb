# Remove data from Data Frames
- axis = 1 -> will remove columns
- axis = 0 -> will remove rowss

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Product": ["Laptop","Mouse","Keyboard","Monitor","HeadtSet"],
    "UnitSold": [12,50,34,20,40],
    "Price": [120000,1300,1500,25000,6000],
    "Discount":[0.10,0.05,0.00,0.12,0.07]
})
```

```python
df
```

```python
df.drop('Discount',axis=1)
```

```python
df.drop(3,axis=0)
```
